대칭 이라는 말은 어떤 원소를 기준으로 양옆이 같을 때 대칭이라고 정의합니다. 이문제에서 대칭 부분의 길이는 무조건 홀수입니다.      

[1,0,1,0,1] => 2번 인덱스 1을 기준으로 뒤집었을때 양쪽이 같음, 대칭 부분의 길이는 5  
2번 인덱스에서 출발  
양옆으로 +1칸 하면 왼쪽은 1번, 오른쪽은 3번, 이 위치를 비교  
양옆으로 +1칸 하면 왼쪽은 0번, 오른쪽은 4번, 이 위치를 비교  
범위를 벗어나면 중단  
왼쪽과 오른쪽이 각각 1,2,3... 만큼 멀어지고 있습니다. 이 멀어지는 크기를 반복문을 통해 나타내보세요.
